```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation # Импортируем FuncAnimation

#  x=np.linspace(0.4,0.7,1000)
#  x0=0.5
#  plt.plot(x,((2**x0**4)*np.cos(20*x0**2+5))+(np.log(2)*2**(x0**4+2)*x0**3*np.cos(20*x0**2+5)-40*2**x0**4*x*np.sin(20*x0**2+5))*(x-x0))
# x=np.linspace(0.2,0.9,1000)
# plt.plot(x,((2**x0**4)*np.cos(20*x0**2+5))-1/((np.log(2)*2**(x0**4+2)*x0**3*np.cos(20*x0**2+5)-40*2**x0**4*x*np.sin(20*x0**2+5)))*(x-x0))
# x=np.linspace(0,1,1000)
# plt.plot(x,(2**x**4)*np.cos(20*x**2+5))
# plt.grid ()
# ax.scatter (1,2)
# ax.scatter (0.920,-1.645)
# plt.subplot(1, 4, 2)
# plt.plot(x,np.log(2)*2**(x**4+2)*x**3*np.cos(20*x**2+5)-40*2**x**4*x*np.sin(20*x**2+5))
# plt.grid ()
# plt.subplot(1, 4, 3)
# plt.plot(x,(np.log(2)*2**(x**4+4)*x**6+3*np.log(2)*2**(x**4+2)*x**2)*np.cos(20*x**2+5)-40*np.log(2)*2**(x**4+2)*x**4*np.sin(20*x**2+5)+(-40*np.log(2)*2**(x**4+2)*x**4-40*2**x**4)*np.sin(20*x**2+5)-800*2**x**2*np.cos(20*x**2+5))
# plt.grid ()
# plt.subplot(1, 4, 4)
# x=np.linspace(0,1,10)
# for x0 in x:
#     plt.plot(x,((2**x0**4)*np.cos(20*x0**2+5))+(np.log(2)*2**(x0**4+2)*x0**3*np.cos(20*x0**2+5)-40*2**x0**4*x*np.sin(20*x0**2+5))*(x-x0))
# x=np.linspace(0,1,1000)
# plt.plot(x,(2**x**4)*np.cos(20*x**2+5))
# plt.grid ()
# plt.show()

def update_cos(frame, line, x):
    # frame - параметр, который меняется от кадра к кадру
    # в данном случае - это начальная фаза (угол)
    # line - ссылка на объект Line2D
    line.set_ydata( (2**(x+frame)**4)*np.cos(20*(x+frame)**2+5) )
    return [line]
fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot(1, 4, 1)
x=np.linspace(0,1,1000)
y=(2**x**4)*np.cos(20*x**2+5)
line, = ax.plot(x,y)

phasa = np.arange(0, 4*np.pi, 0.1) 
animation = FuncAnimation(
    fig,                # фигура, где отображается анимация
    func=update_cos,    # функция обновления текущего кадра
    frames=phasa,       # параметр, меняющийся от кадра к кадру
    fargs=(line, x),    # дополнительные параметры для функции update_cos
    interval=30,       # задержка между кадрами в мс
    blit=True,          # использовать ли двойную буферизацию
    repeat=False)       # зацикливать ли анимацию
animation.save('cos.gif', writer='imagemagick')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation # Импортируем FuncAnimation
from matplotlib.animation import ArtistAnimation

#  x=np.linspace(0.4,0.7,1000)
#  x0=0.5
#  plt.plot(x,((2**x0**4)*np.cos(20*x0**2+5))+(np.log(2)*2**(x0**4+2)*x0**3*np.cos(20*x0**2+5)-40*2**x0**4*x*np.sin(20*x0**2+5))*(x-x0))
# x=np.linspace(0.2,0.9,1000)
# plt.plot(x,((2**x0**4)*np.cos(20*x0**2+5))-1/((np.log(2)*2**(x0**4+2)*x0**3*np.cos(20*x0**2+5)-40*2**x0**4*x*np.sin(20*x0**2+5)))*(x-x0))
# x=np.linspace(0,1,1000)
# plt.plot(x,(2**x**4)*np.cos(20*x**2+5))
# plt.grid ()
# ax.scatter (1,2)
# ax.scatter (0.920,-1.645)
# plt.subplot(1, 4, 2)
# plt.plot(x,np.log(2)*2**(x**4+2)*x**3*np.cos(20*x**2+5)-40*2**x**4*x*np.sin(20*x**2+5))
# plt.grid ()
# plt.subplot(1, 4, 3)
# plt.plot(x,(np.log(2)*2**(x**4+4)*x**6+3*np.log(2)*2**(x**4+2)*x**2)*np.cos(20*x**2+5)-40*np.log(2)*2**(x**4+2)*x**4*np.sin(20*x**2+5)+(-40*np.log(2)*2**(x**4+2)*x**4-40*2**x**4)*np.sin(20*x**2+5)-800*2**x**2*np.cos(20*x**2+5))
# plt.grid ()
# plt.subplot(1, 4, 4)
# x=np.linspace(0,1,10)
# for x0 in x:
#     plt.plot(x,((2**x0**4)*np.cos(20*x0**2+5))+(np.log(2)*2**(x0**4+2)*x0**3*np.cos(20*x0**2+5)-40*2**x0**4*x*np.sin(20*x0**2+5))*(x-x0))
# x=np.linspace(0,1,1000)
# plt.plot(x,(2**x**4)*np.cos(20*x**2+5))
# plt.grid ()
# plt.show()

fig = plt.figure(figsize=(10, 6))
ax_3d = fig.add_subplot(projection='3d')
x = np.arange(-2*np.pi, 2*np.pi, 0.2)
y = np.arange(-2*np.pi, 2*np.pi, 0.2)
xgrid, ygrid = np.meshgrid(x, y)
phasa = np.arange(0, 2*np.pi, 0.1)
frames = []
line, = ax_3d.plot(x,y)

for p in phasa:
    zgrid = np.sin(xgrid+p) * np.sin(ygrid) / (xgrid * ygrid)
 
    line = ax_3d.plot_surface(xgrid, ygrid, zgrid, color='b')
    frames.append([line])
animation = ArtistAnimation(
    fig,                # фигура, где отображается анимация
    frames,              # кадры
    interval=30,        # задержка между кадрами в мс
    blit=True,          # использовать ли двойную буферизацию
    repeat=True)         # зацикливать ли анимацию
animation.save('cos.gif', writer='imagemagick')
plt.show()
```
